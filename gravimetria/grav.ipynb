## GRUPO-I

Afortunadamente en este grupo se tomaron coordenadas GPS con un dispositivo móvil telefónico. Este grupo hizo dos lineas de Survey, por lo que los datos al final se separar en dos, después de las correcciones necesarias aplicadas a la gravedad. Las coordenadas nos dejarán crear una gráfica en 3D de los valores de gravedad luego de las correcciones.

```python
#Para el Grupo I
datos1=loadtxt('grav_grupoI.txt')
datos2=loadtxt('grav_grupoI2.txt')
ro=2461
G=6.67*10**(-11)
gfa=0.3086
grav=[]
tiempo=[]
alt=[]
x=[]
y=[]
for i in range(0,len(datos1)):
    grav.append((datos1[i,3]+datos2[i,3])/2)
    tiempo.append(((datos1[i,11]*3600+datos1[i,12]*60+datos1[i,13])+(datos2[i,11]*3600+datos2[i,12]*60+datos2[i,13]))/2)
    alt.append((datos1[i,18]+datos2[i,18])/2)
    y.append((datos1[i,17]+datos2[i,17])/2)
    x.append((datos1[i,16]+datos2[i,16])/2)
m=((grav[24]-grav[0])/(tiempo[24]-tiempo[0]))
gdrift=[]
temp=tiempo[0]
for i in range(0,len(tiempo)):
    tiempo[i]=tiempo[i]-temp
    gdrift.append(grav[i]-m*tiempo[i])
galt=[]
for i in range(0,len(gdrift)):
    galt.append(gdrift[i]+(gfa*alt[i]))
gbouguer=[]
for i in range(0,len(galt)):
    gbouguer.append(galt[i]-(2*pi*G*alt[i]*ro))
```

```python
gfinal=[]
altfinal=[]
xfinal=[]
yfinal=[]
for i in range(0,len(gbouguer)):
    if alt[i]!=0:
        gfinal.append(gbouguer[i])
        altfinal.append(alt[i])
        xfinal.append(x[i])
        yfinal.append(y[i])
gravL1=[]
altL1=[]
gravL2=[]
altL2=[]
xL1=[]
yL1=[]
xL2=[]
yL2=[]
for i in range(0,len(gfinal)-11):
    gravL1.append(gfinal[i])
    altL1.append(altfinal[i])
    xL1.append(xfinal[i])
    yL1.append(yfinal[i])
for i in range(0,len(gfinal)-9):
    gravL2.append(gfinal[9+i])
    altL2.append(altfinal[9+i])
    xL2.append(xfinal[9+i])
    yL2.append(yfinal[9+i])

baseL1=gravL1[0]
baseL2=gravL2[0]
for i in range(0,len(gravL1)):
    gravL1[i]=gravL1[i]-baseL1
for i in range(0,len(gravL2)):
    gravL2[i]=gravL2[i]-baseL2
```

```python
scatter(altL1,gravL1)
xlabel('Altua(m)')
ylabel('Gravedad(mGal)')
title('Linea 1')
```

```python
scatter(altL2,gravL2)
xlabel('Altua(m)')
ylabel('Gravedad(mGal)')
title('Linea 2')
```

## GRUPO A

Este grupo en particular no tiene un medio ideoneamente homogeneo en algún tipo de roca. De la estación 1 a la 5 están en un medio de  arena, del 6 al 13 en volcánicas (Dacita), y de la 14 en adelante en basamento metamórfico (Esquisto).

```python
#Para el Grupo A
datos=loadtxt('grav_grupoA.txt')
ro1=1750
ro2=2400
ro3=2900
G=6.67*10**(-11)
gfa=0.3086
grav=[]
tiempo=[]
alt=[]
x=[]
y=[]
for i in range(0,len(datos)):
    grav.append(datos[i,3])
    tiempo.append(datos[i,11])
    alt.append(datos[i,2])
    #y.append((datos1[i,17]+datos2[i,17])/2)
    #x.append((datos1[i,16]+datos2[i,16])/2)
m=((grav[len(datos)-1]-grav[0])/(tiempo[len(datos)-1]-tiempo[0]))
gdrift=[]
temp=tiempo[0]
for i in range(0,len(tiempo)):
    tiempo[i]=tiempo[i]-temp
    gdrift.append(grav[i]-m*tiempo[i])
galt=[]
for i in range(0,len(gdrift)):
    galt.append(gdrift[i]+(gfa*alt[i]))
gbouguer=[]
for i in range(0,len(galt)):
    if i<=4 and i>=0:
        gbouguer.append(galt[i]-(2*pi*G*alt[i]*ro1))
    if i<=12 and i>=5:
        gbouguer.append(galt[i]-(2*pi*G*alt[i]*ro2))
    if i>=13:
        gbouguer.append(galt[i]-(2*pi*G*alt[i]*ro3))
```

```python
anomaly=[]
base=gbouguer[0]
for i in range(0,len(galt)):
    anomaly.append(gbouguer[i]-base)
```

```python
plot(tiempo,anomaly)
xlabel('Tiempo (m)')
ylabel('Anomalia de Gravedad (mGal)')
title('Linea Grupo A')
```
