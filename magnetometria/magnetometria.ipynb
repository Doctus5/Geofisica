```python
%pylab inline
from mpl_toolkits.mplot3d import Axes3D
```

```python
X=[]
Y=[]
Z=[]
```

```python
def new_matrix(p):
    for i in range(0,len(p)):
        X.append(p[i][0])
        Y.append(p[i][1])
        Z.append(p[i][2])
```

```python
def fix(p):
    for i in range(0,len(p)):
        x=p[i][0]
        y=p[i][1]
        for j in range(0,len(p)):
            X=p[j][0]
            Y=p[j][1]
            if x==X and y==Y:
                Z[i]=(p[i][2]+p[j][2])/2.0
```

```python
work=loadtxt("DatosMagnetometria_GrupoH2.csv",usecols=(0,1,2),delimiter=',')
X=[]
Y=[]
Z=[]
new_matrix(work)
fix(work)
plot(X,Y)
xlabel(r'Desplazamiento X', fontsize=15)
ylabel(r'Desplazamiento Y', fontsize=15)
title(r'Ruta de Medicion de Grupo H',fontsize=20)
fig = figure()

fig = figure(figsize=(16,8))
ax = fig.gca(projection='3d')
p=ax.plot_trisurf(X,Y,Z, cmap=cm.jet, linewidth=0.2)
c2 = fig.colorbar(p)
ax.set_xlabel(r'Desplazamiento X', fontsize=15)
ax.set_ylabel(r'Desplazamiento Y', fontsize=15)
ax.set_zlabel(r'nT', fontsize=15)
ax.set_title(r'Campo Magnetico de Grupo H',fontsize=25)
```
