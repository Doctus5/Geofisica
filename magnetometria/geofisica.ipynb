```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python

A=np.loadtxt("DatosMagnetometria_GrupoH2.csv",usecols=(0,1,2),delimiter=",")
x=A[:,0]
y=A[:,1]
z=A[:,2]
z1=[]
x1=[]
y1=[]
for i in range(0,size(z)):
    for j in range (0,size(z)):
        if (x[i]==x[j]and y[i]==y[j] and i!=j ):
            prom=(z[i]+z[j])/2
            z1.append(prom)
            x1.append(x[i])
            y1.append(y[i])
        
                       
print size(z1)
print size(z)
plot(x,y)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x,y,z)

fig = plt.figure(figsize=(16,8))
ax = fig.gca(projection='3d')
p=ax.plot_trisurf(x1,y1,z1, cmap=cm.jet, linewidth=0.2)
c2 = fig.colorbar(p)
ax.set_xlabel(r'Eje X', fontsize=20)
ax.set_ylabel(r'Eje Y', fontsize=20)
ax.set_zlabel(r'nanoTeslas (nT)', fontsize=20)
ax.set_title(r'Induccion magnetica ',fontsize=25)
```
